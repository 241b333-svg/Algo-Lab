```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
x=print(func1(2,4))
import time
start = time.time()
f=func2(2,4)
print(f)
end= time.time()
totaltime=(end-start)*1000
print(totaltime)
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
    return result
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
x=print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
import time
start = time.time()
f=func4([1, 2, 3, 4], [2, 4, 6, 8])
print(f)
end= time.time()
totaltime=(end-start)*1000
print(totaltime)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
x=print(func5(16))
import time
start = time.time()
f=func5(16)
print(f)
end= time.time()
totaltime=(end-start)*1000
print(totaltime)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
x=print(func8([3,1,2,7,8,6]))
import time
start=time.time()
f=func8([3,1,2,7,8,6])
print(f)
end=time.time()
totaltime=(end-start)*1000
print(totaltime)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total

numbers = [1, 2, 3, 4, 5]
print("Sum of array:", sum_arr(numbers))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr, find_max=True):
    if not arr:  # handle empty array
        return None

    result = arr[0]
    for num in arr[1:]:
        if find_max and num > result:
            result = num
        elif not find_max and num < result:
            result = num
    return result

numbers = [12, 3, 4, 15, 45, 78, 450]

print("Maximum:", max_min(numbers, find_max=True))
print("Minimum:", max_min(numbers, find_max=False))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count
numbers = [12, 1234, 45, 67, 1]
even, odd = Counting_EvenOdd(numbers)
print("Even count:", even)
print("Odd count:", odd)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

numbers = [12, 1234, 45, 67, 1]
target = 45
result = linear_Search(numbers, target)

if result != -1:
    print(f"Element {target} found at index {result}")
else:
    print(f"Element {target} not found in array")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1, 3, 5, 7, 9, 11, 13]
target = 7
result = Binary_Search(arr, target)

if result != -1:
    print(f"Element {target} found at index {result}")
else:
    print(f"Element {target} not found in array")
```
